# Generate the Figures from 1812.06976

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['font.serif'] = 'CMU Serif'
```

# Figure 1

```python
# Update figure parameters
mpl.rcParams['figure.figsize'] = 20, 16
mpl.rcParams['xtick.top'] = True
mpl.rcParams['xtick.major.size'] = 10.
mpl.rcParams['xtick.minor.size'] = 5.
mpl.rcParams['xtick.major.width'] = 2.0
mpl.rcParams['xtick.minor.width'] = 1.5
mpl.rcParams['xtick.major.pad'] = 16.
mpl.rcParams['xtick.labelsize'] = 44.
mpl.rcParams['xtick.direction'] = 'in'
mpl.rcParams['xtick.minor.visible'] = True
mpl.rcParams['ytick.right'] = True
mpl.rcParams['ytick.major.size'] = 10.
mpl.rcParams['ytick.minor.size'] = 5.
mpl.rcParams['ytick.major.width'] = 2.0
mpl.rcParams['ytick.minor.width'] = 1.5
mpl.rcParams['ytick.major.pad'] = 16.
mpl.rcParams['ytick.labelsize'] = 44.
mpl.rcParams['ytick.direction'] = 'in'
mpl.rcParams['ytick.minor.visible'] = True
mpl.rcParams['lines.dashed_pattern'] = 4.5, 3.

# Load Fig 1 data
lim_data = np.loadtxt('./data/Fig1data.txt')

fig = plt.figure()
ax = fig.add_subplot(111)
plt.yscale('log')
plt.xlabel(r"$m_s $ [keV]",fontsize=55)
plt.ylabel(r"$\sin^2(2 \theta)$",fontsize=55)

plt.axis([6.7,7.4,2.e-13,6.e-9]) 

# Expectation
plt.fill_between(lim_data[0],lim_data[4],lim_data[5],color="gold",alpha=0.95)
plt.fill_between(lim_data[0],lim_data[3],lim_data[4],color="limegreen",alpha=0.8)
plt.plot(lim_data[0],lim_data[2],color="black",linestyle="dashed",label=r"MC: median",lw=2.)

# Limit
plt.plot(lim_data[0],lim_data[1],c='black',label="95\% data limit (this work)",linewidth=3.5,marker='s',markersize=12)

# Legend
m3, = plt.plot([], [], c='limegreen' , marker='s', markersize=40,fillstyle='left', linestyle='none',alpha=0.8)
m4, = plt.plot([], [], c='gold' , marker='s', markersize=40,fillstyle='right', linestyle='none',alpha=0.95)
handles,labels = ax.get_legend_handles_labels()
handles = [handles[1],handles[0],(m3,m4)]
labels = [u'95\% limit (this work)',
          u'mean expected',
          u'1$\sigma$/2$\sigma$ containment']
axbox = ax.get_position()
x_value=-0.12
y_value= -.12
leg = plt.legend(handles,labels,loc = (axbox.x0 + x_value, axbox.y0 + y_value),fontsize=40,fancybox=True,facecolor='none',frameon=False)

plt.tight_layout()
plt.show()
```

# Figure 2

```python
# Update plotting parameters
mpl.rcParams['figure.figsize'] = 8, 8
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.minor.size'] = 2.5
mpl.rcParams['xtick.major.width'] = 1.0
mpl.rcParams['xtick.minor.width'] = 0.75
mpl.rcParams['xtick.major.pad'] = 8
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.minor.size'] = 2.5
mpl.rcParams['ytick.major.width'] = 1.0
mpl.rcParams['ytick.minor.width'] = 0.75
mpl.rcParams['ytick.major.pad'] = 8
mpl.rcParams['legend.frameon'] = True
mpl.rcParams['lines.dashed_pattern'] = 2.8, 1.5

# Load Fig 2 data
ms_mdl = np.loadtxt('./data/Fig2_ms_mdl.txt')
ms_dat = np.loadtxt('./data/Fig2_ms_dat.txt')
pn_mdl = np.loadtxt('./data/Fig2_pn_mdl.txt')
pn_dat = np.loadtxt('./data/Fig2_pn_dat.txt')

fig = plt.figure()

# MOS
ax1 = fig.add_axes([0.1, 0.5, 1.5, 0.4],xticklabels=[],ylim=(0.079,0.092))
ax1.errorbar(ms_dat[0],ms_dat[1],xerr=ms_dat[2],yerr=ms_dat[3],color="black",label="data",marker="o", fmt='none',capsize=4)
ax1.plot(ms_mdl[0],ms_mdl[1],'r-',label="back. model",lw=2.)
ax1.plot(ms_mdl[0],ms_mdl[2],'r--',label="back. + signal",lw=2.)

# PN
ax2 = fig.add_axes([0.1, 0.1, 1.5, 0.4],ylim=(0.096,0.113))
ax2.errorbar(pn_dat[0],pn_dat[1],xerr=pn_dat[2],yerr=pn_dat[3],color="black",label="data",marker="o", fmt='none',capsize=4)
ax2.plot(pn_mdl[0],pn_mdl[1],'r-',label="back. model",lw=2.)
ax2.plot(pn_mdl[0],pn_mdl[2],'r--',label="back. + signal",lw=2.)
ax2.legend(ncol=3,loc=2,fontsize=28,frameon=False)

plt.xlabel(r"$E_{\rm line}$ [keV]",fontsize=28)
plt.ylabel(r"counts/s/keV",fontsize=28,alpha=0.)
ax1.text(3.24,0.083,r"Flux [counts/s/keV]",fontsize=28,rotation=90)
ax1.text(3.71,0.0902,r"\textbf{MOS data}",fontsize=28)
ax2.text(3.725,0.1105,r"\textbf{PN data}",fontsize=28)
ax1.text(3.31,0.080,r"\textbf{A}",fontsize=28)
ax2.text(3.31,0.097,r"\textbf{B}",fontsize=28)

ax1.set_xlim(3.3,3.8)
ax2.set_xlim(3.3,3.8)
ax1.xaxis.set_tick_params(labelsize=28)
ax2.xaxis.set_tick_params(labelsize=28)
ax1.yaxis.set_tick_params(labelsize=28)
ax2.yaxis.set_tick_params(labelsize=28)

plt.tight_layout()
plt.show()
```

# Figure 3

```python
# Update plotting parameters
mpl.rcParams['axes.labelsize'] = 22
mpl.rcParams['xtick.labelsize'] = 22
mpl.rcParams['ytick.labelsize'] = 22

# Load Fig 3 data
TS = np.loadtxt('./data/Fig3_TS.txt')
chisq_dat = np.loadtxt('./data/Fig3_chisq_dat.txt')
chisq_mdl = np.loadtxt('./data/Fig3_chisq_mdl.txt')

fig, ax1 = plt.subplots(figsize=(10,8))
ax2 = fig.add_axes([0.41, 0.43, 0.45, 0.4])

ax1.fill_between(TS[0],np.zeros(len(TS[0])),1*np.ones(len(TS[0])),color="limegreen",alpha=0.8)
ax1.fill_between(TS[0],np.ones(len(TS[0])),4*np.ones(len(TS[0])),color="gold",alpha=0.95)

ax1.plot(TS[0],TS[1],'k-',linewidth=2.0)

ax1.set_xlim(6.7,7.4)
ax1.set_xlabel(r"$m_s$ [keV]",fontsize=22)
ax1.set_ylabel(r"TS",fontsize=22)
ax1.set_ylim(0,5)

ax2.errorbar(chisq_dat[0],chisq_dat[1],xerr=chisq_dat[2],yerr=(chisq_dat[3],chisq_dat[4]),
             marker="o", fmt='none',capsize=5,color="black",label="data")
ax2.plot(chisq_mdl[0],chisq_mdl[1],'r--',label="$\chi^2$ distribution")
ax2.set_yscale('log')
ax2.set_xlim(0.,12)
ax2.set_ylim(0.1,3e2)

ax2.set_xlabel(r"TS",fontsize=22)
ax2.set_ylabel(r"$dN_{\rm obs}/d({\rm TS})$",fontsize=22)

ax1.text(6.72,4.6,r"\textbf{A}",fontsize=22*1.2)
ax2.text(0.5,0.2,r"\textbf{B}",fontsize=22*1.2)

plt.show()
```
